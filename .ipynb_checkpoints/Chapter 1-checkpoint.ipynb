## Вопрос 1
***

*Какое из приведенных ниже выражений соответствует высказыванию: "Вероятность того, что 9 июля 1816 года был солнечный день"?*
- p(солнечно)
- p(солнечно|июль)
- p(солнечно|9 июля 1816 года)
- p(9 июля 1816 года|солнечно)
- p(солнечно, 9 июля 1816 года)/p(9 июля 1816 года)


__Ответ:__

Формула условной вероятности:

$$ p(A|B) $$

"Читается, что если наступило событие A при условии наступления события B"

Тогда нам подходит следующее выражение:

$$ p(солнечно|9 июля 1816 года) $$

Однако рассмотрим формулу условной вероятности:

$$ p(A|B) = \frac{p(A,B)}{p(B)} $$

Отсюда следует, что также годится и следующая запись:

$$ p(солнечно, 9 июля 1816 года) / p(9 июля 1816 года) $$

не равно вероятности того, Папа является человеком.*

__Ответ:__


Вероятность того, что 1 человек из 7 млрд. людей является Папой Римским:

$$ p(Папа|человек)=\frac{1}{7 000 000 000} $$

Вероятность того, что Папа Римский является человеком:

$$ p(человек|Папа)=1 $$

## Вопрос 5
***

*Записать теорему Байеса для модели из упражнения 3.*

__Ответ:__

Теорема Байеса:

$$ p(\theta|y)=\frac{p(y|\theta)p(\theta)}{p(y)}, $$

где $\theta$ - предположение, $p(\theta)$ - априорная вероятность.

Для модели из упражнения 3 $\theta = \mu, \sigma$, следовательно, $p(\theta)=p(\mu,\sigma)=p(\mu)p(\sigma)$, так как они независимы. 

Тогда можно окончательно переписать:

$$ p(\mu, \sigma)=\frac{p(y|\mu, \sigma)p(\mu)p(\sigma)}{p(y)}$$

## Вопрос 6
***

*Предположим, что есть две монеты. При подбрасывании первой монеты в половине случаев выпдают решки, в половине случаев - орлы. Вторая монета несимметричная ("мошенническая") - при ее подбрасывании всегда выпадают орлы. Если наугад взять одну из монет и при подбрасывании получить орел, то какова вероятность того, что эта монета евляется несимметричной?*

__Ответ:__

Формализуем постановку задачи. Пусть событие О - выпадение орла, Р - выпадение решки, С - монета симметричная, Н - монета не симметричная.

Тогда по условию задачи нам дано:
\begin{eqnarray}
p(О | С) = p(Р | C) = 0.5 \newline
p(O | H) = 1,\space p(P | H) = 0
\end{eqnarray}

Также известно, что $p(C) = p(H) = 0.5$.

Нам надо найти $p(H | O)$.

Запишем для этого выражения теорему Байеса:

$$ p(H | O) = \frac{p(O | H)p(H)}{p(O)} $$

Для этого нам надо записать лишь вероятность выпадения орла. По формуле полной вероятности:

$$ p(O) = p(O | H) * p(H) + p(O | C) * p(C) $$

Тогда окончательно формула будет выглядеть следующим образом:

$$ p(H | O) = \frac{p(O | H)p(H)}{p(O | H) * p(H) + p(O | C) * p(C)} $$

```python
1 * 0.5 / (1 * 0.5 + 0.5 * 0.5)
```

## Вопрос 7 и 8
***

*Добавить пунктирную вертикальную линию, показывающую отношение орел/(количество бросков).*

*Проварьировать априорные вероятности и другие данные.*

```python
plt.figure(figsize=(10, 8))

n_trials = [0, 1, 2, 3, 4, 8, 16, 32, 50, 150]
data = [0, 1, 1, 1, 1, 4, 6, 9, 13, 48]
theta_real = 0.35

beta_params = [(1, 1), (20, 20), (1, 4)]
dist = stats.beta
x = np.linspace(0, 1, 200)

for idx, N in enumerate(n_trials):
    if idx == 0:
        plt.subplot(4, 3, 2)
        plt.xlabel(r'$\theta$')
    else:
        plt.subplot(4, 3, idx+3)
        plt.xticks([])
    y = data[idx]
    for (a_prior, b_prior) in beta_params:
        p_theta_given_y = dist.pdf(x, a_prior + y, b_prior + N - y)
        plt.fill_between(x, 0, p_theta_given_y, alpha=0.7)
        
        # Пунктирная линия, показывающая отношение орел/(количество бросков)
        try:
            unit_rate_per_toss = y/N
        except ZeroDivisionError:
            unit_rate_per_toss = 0
        plt.axvline(unit_rate_per_toss, ymax=0.3, 
                    color='k', linestyle="--")       
        
    plt.axvline(theta_real, ymax=0.3, color='k')
    plt.plot(0, 0, label=f'{N:4d} trials\n{y:4d} heads', alpha=0)
    plt.xlim(0, 1)
    plt.ylim(0, 12)
    plt.legend()
    plt.yticks([])
plt.tight_layout()
```

Синий график - априорное распределение, которое представляет все возможные значения для отклонения, вероятно равного апрриори.

Оранжевый - гаусс-подобное априорное распределение сконцентрированное в центральной области около значения 0,5.

Зеленый - ассимитричное априорное распределение, которое придает наибольший вес результатам с отклонением в сторону выпадения решки.
