```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc3 as pm

from scipy import stats

np.random.seed(123)
```

### Вопрос 1
***

*Используя библиотеку PyMC3, изменить параметры априорного распределения в модели  `our_first_model`, чтобы они соответствовали параметрам из параметра в предыдущей главе. Сравнить полученные результаты с результатом из предыдущей главы. Заменить бета-распределение на равномерное в интервале [0,1]. Равнозначны ли полученные результаты результатам для $Beta(\alpha=1,\beta=1)$? Выполняется ли выборка быстрее, медленнее или за то же время? Как измениться результат при использовании более широкого интервала, например [-1,2]? Остается ли модель работоспособной? Какие сообщения об ошибках выводятся?*

__Ответ__:

#### Основная модель

Основная модель с предварительными значениями $\alpha=1,\beta=1$.

При сравнении модели будем пользоваться следующей стретегией:

1. Запустить модель.
2. Прочитать предупреждения.
3. Посмотреть апостериорную оценку плотности и график трассировки.
4. Изучить числовую сводку.

```python
trials = 4
theta_real = 0.35 #в реальном эксперименте это знчение неизвестно
data = stats.bernoulli.rvs(p=theta_real, size=trials)
```

```python
with pm.Model() as our_first_model:
    θ = pm.Beta('θ Beta(alpha=1,beta=1)', alpha=1., beta=1.)
    y = pm.Bernoulli('y', p=θ, observed=data)
    base_trace = pm.sample(10000, random_seed=123)
```

```python
az.plot_trace(base_trace)
```

```python
base_model = az.summary(base_trace)
base_model
```

#### Пункт 1: Сравнение с предыдущим примером

```python
with pm.Model() as our_model:
    θ = pm.Beta("θ", alpha=1, beta=1)
    y = pm.Bernoulli("y", p=θ, observed=data)
    trace = pm.sample(1000, random_seed=123)
```

```python
az.plot_trace(trace)
```

```python
az.summary(trace)
```

#### Пункт 2: Замена бета-распределение на равномерное в интервале [0,1]

```python
with pm.Model() as our_model:
    θ = pm.Uniform("θ Uniform(0,1)", lower=0, upper=1)
    y = pm.Bernoulli("y", p=θ, observed=data)
    trace_uniform_0_1 = pm.sample(10000, random_seed=123)
```

```python
az.plot_trace(trace_uniform_0_1)
```

```python
uniform_0_1_model = az.summary(trace_uniform_0_1)
pd.concat([base_model, uniform_0_1_model])
```

#### Пункт 3: Замена на отрезок [-1,2]

```python
with pm.Model() as our_model:
    θ = pm.Uniform("θ Uniform(-1,2)", lower=-1, upper=2)
    y = pm.Bernoulli("y", p=θ, observed=data)
    trace_uniform_1_2 = pm.sample(10000, random_seed=123)
```

```python
az.plot_trace(trace_uniform_1_2)
```

```python
uniform_1_2_model = az.summary(trace_uniform_1_2)
pd.concat([base_model, uniform_1_2_model])
```

Что поменялось: 

* Появились предупреждения `There were 1467 divergences after tuning. Increase 'target_accept' or reparameterize`. С выборкой что-то не так.

* Появились черные полосы под графиком.

Можно видеть, что упала и эффективность количества выборок, до этого в среднем было 17126, стало - 12953. Это означает, что полезных образов стало меньше, чем общее количество образцов.
